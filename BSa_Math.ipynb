```python
def loc_location(winnings):
  prob = 0.33
  EV = []
  for i in range(len(winnings)-5):
    p = i + 4
    calc =  ( 1 - prob**p) * winnings[i] + prob**p *winnings[p]
    EV.append(calc)
  return EV
```

```python
loc_locations = loc_location(winnings)
```
