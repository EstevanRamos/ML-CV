```python
import numpy as np
from sklearn.cluster import KMeans
```

```python
def RandIndex(known, result):
  TP,FP,TN,FN = 0
  for i in range(0,len(result)-1):
    for j in range(i+1 , len(known)):
      if result[i] == result[j]:
        if known[i] == known[j]:
          TP = TP+1
        else:
          FP = FP + 1
      else:
        if known[i] != known[j]:
          TN = TN + 1
        else:
          FN = FN + 1
    return (TP+TN)/(TP+TN+FP+FN)
```

```python
def RI(known,result):
  TP,FP,TN,FN = 0
  TP = np.sum([result[:len(result)-1] == result[1:] & known[:len(known)-1] == known[1:]])
  FP = np.sum([result[:len(result)-1] == result[1:] & known[:len(known)-1] != known[1:]])
  TN = np.sum([result[:len(result)-1] != result[1:] & known[:len(known)-1] != known[1:]])
  FN = np.sum([result[:len(result)-1] != result[1:] & known[:len(known)-1] == known[1:]])
  return (TP+TN)/(TP+TN+FP+FN)
```

```python
def purity(assignment, known):
  aLabels = set(assignment)
  alabels = list(alabels)
  kLabels = set(known)
  klabels = list(klabels)

  maxOverlaps = []
  for cID in aLabels:
    indiciesOfCID = [ii for ii in range(0,len(assignment)) if assignment[ii] == cID]
    overlap = []

    for kID in kLabels:
      indiciesOfKID = [ii for ii in range(0,len(known)) if known[ii] == kID]
      overlap.append(len(set(indiciesOfCID).intersection(indeciesOfKID)))

    maxOverLaps.append(max(overlap))
  p = sum(maxOverlaps)/len(assignment)
  return p
```

```python
d1 = np.random.rand(10,2)
d2 = 2 + np.random.rand(10,2)
d3 = 5 + np.random.rand(10,2)
d4 = np.random.rand(10,2)

data = np.append(d1,d2, axis = 0)
data = np.append(data,d3, axis = 0)
data = np.append(data,d4, axis = 0)

k = 3

knn = KMeans(n_clusters = k).fit(data)
print(knn.labels_)
print(knn.cluster_centers_)
print(knn.inertia_)
```
