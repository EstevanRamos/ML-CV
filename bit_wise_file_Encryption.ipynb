```python
def bitwise_or_and_write(file_path, output_file_path, bitwise):
  b = bitwise
  print('1')
  newf = open(output_file_path,'wb')
  print('2')
  with open(file_path, 'rb') as f:
    print('3')
    while(byte := f.read(1)):
      newbyte = bytes(a ^ b for a, b in zip(byte, b))
      newf.write(newbyte)
  newf.close()
```

```python
# Example usage:
input_file_path = '/content/un-encrypted.cap'
output_file_path = '/content/encrypted.cap'
byte = b'\x14'
bitwise_or_and_write(input_file_path, output_file_path,byte)
print("Result written to output file:", output_file_path)
```
