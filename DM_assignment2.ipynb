```python
def RandIndex( known, result):
    TP=0
    FP=0
    TN=0
    FN=0
    
    for i in range(0, len(result)-1):
        for j in range( i+1, len(result)):
            if result[i]==result[j]:
                # Positive
                if known[i]==known[j]:
                    # True
                    TP=TP+1
                else: #False
                    FP=FP+1
            else: #Negative
                if known[i]!=known[j]:
                    #True
                    TN=TN+1;
                else:#False
                    FN=FN+1
    RI = (TP+TN)/(TP+TN+FP+FN)
    return RI


def purity(assignment, known):
    aLabels = set(assignment)
    aLabels = list(aLabels)
    kLabels = set(known)
    kLabels = list(kLabels)
    
    maxOverlaps=[]
    for cID in aLabels:
        indicesOfCID= [ii for ii in\
            range(0, len(assignment))\
            if assignment[ii]==cID]
        overlap=[]
        
        for kID in kLabels:
            indicesOfKID= [ii for ii in\
                           range(0, len(known))\
                               if known[ii]==kID]
            
            overlap.append(\
                len(set(indicesOfCID).intersection\
                (indicesOfKID))\
                    )
                        
        maxOverlaps.append(max(overlap))
            
    p = sum(maxOverlaps)/len(assignment)
    return p
```

```python
d = pd.read_csv('dry_beans_data.csv')
t = pd.read_csv('dry_beans_benchmarkclusters.csv', header = None)
```

```python
data = d.to_numpy()
truth = t.values.reshape(-1)
```

```python
print(data.shape)
print(t.shape)
```

```python
 kmeans = KMeans(n_clusters=7).fit(data)
```

```python
 occurunces = []
 for i in range(7):
   occurunces.append(np.sum([kmeans.labels_ == i]))
print(np.sum(occurunces))
```

```python
print(RandIndex(truth,kmeans.labels_))
print(purity(truth,kmeans.labels_))
```

```python
asc=silhouette_score(data, kmeans.labels_)
```
