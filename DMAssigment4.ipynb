```python
import numpy as np
beta = 0.8
threshold = 0.00000000000001

#Yahoo, amazon, msoft


A=[[1, 1, 0],
   [1, 0, 1],
   [0, 1, 0]]

arr= np.array(A, dtype=float)

s=[]
for i in range(0, len(A)):
    s.append(np.sum(arr[:,i]))

# s contains column-wise summations

M=arr

for j in range(0, len(A)):
    M[:,j]=M[:,j]/s[j]


r = (1.0+np.zeros([len(M), 1]))/len(M)

uniD = (1.0-beta)*r
print(uniD)
rprev = r

for i in range(0, 1000):
    r= beta*np.matmul(M, rprev)+uniD
    
    diff = sum(abs(r-rprev))[0]
    #print(i, " ", diff)
    
    if diff<=threshold:
        print("Valur of i=", i)
        break        
    rprev=r
    
print(r)
```

```python
def p(beta, fig):
  threshold = 0.00000000000001
  fg = np.array(fig, dtype=float)
  x = np.sum(fg,axis=0)
  M = np.divide(fg,x)

  R = np.ones_like(M[0])
  R = R * 1/len(M)
  R = R[:, np.newaxis]
  uniD = (1.0-beta)*R

  for i in range(1000):
    rprev = R
    R = beta*np.dot(M, rprev)+uniD
    diff = sum(abs(R-rprev))[0]
    if diff<=threshold:
      print("Value of i=", i)
      break

  return R 
```
