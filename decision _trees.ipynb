```python
def predicty(x , y, t):
  return np.mean([x >= t] == y) # returns the accuracy of x and y
```

```python
def predicty(x, y, t=None):
  t = np.mean(x) if not t else t
  return np.mean([x >= t] == y)
```

```python
def randpredicty1(x, y, n):
  mn = np.min(x)
  mx = np.max(x)
  t = np.random.uniform(mn,mx,n)
  results = [x <= i for i in t]
  k = np.mean(results == y , axis = 1)
  ind = np.argmax(k)
  return k[ind] , t[ind]
```

```python
def randpredicty(x, y, n):
  t = (max(x) - min(x)) * np.random.random_sample(n) + min(x)
  best_threshold = None
  best_accuracy = 0.0

  for thresh in t:
    accuracy = np.mean([x <= thresh] == y)

    if accuracy >= best_accuracy:
      best_threshold = thresh
      best_accuracy = accuracy
  
  return best_threshold, best_accuracy
```
